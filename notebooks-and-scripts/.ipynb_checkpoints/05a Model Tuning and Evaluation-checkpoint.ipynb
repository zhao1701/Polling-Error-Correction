# Tuning parameters
The conceptually simplest method of finding the appropriate hyperparameters for a model is to simply test all possible combinations of hyperparameter values within a specified range. For example, for parameters $x \in X = \{1, 2\}$ and $y \in Y = \{a,b,c\}$, the cartesian product $X \times Y = \{(1,a),(1,b),(1,c),(2,a),(2,b),(2,c)\}$ represents all combinations of hyperparameter values to be tested.

Scikit-learn provides an easy interface for testing these combinations called "grid search." As with before, we evaluate the effectiveness of a model with chosen hyperparameters using a cross-validated F1-score due to the unbalanced nature of the data.

### Logistic regression
For logistic regression, we must choose values for *penalty* and *$C$*. A brief overview:
- **$C$**: The inverse of the regularization strength. In a linear model, [Regularization](https://datanice.github.io/machine-learning-101-what-is-regularization-interactive.html) is used to prevent a model from overfitting its data by adding a penalty to the cost function that discourages coefficients from having too large an absolute value. Because $C$ is the inverse of regularization strength, this means that higher values of $C$ result in more complex models more prone to overfitting, and lower values of $C$ result in simpler models more prone to underfitting.
- **penalty**: Specifies the norm used in calculating the regularization penalty component of the cost function. The $l_2$ norm (often referred to as the Euclidian norm) is the most commonly used. For a vector $x \in \mathbb{R}^n$, the $l_2$ norm is defined as $\|x\| = \sqrt{\sum_{i=1}^n x_i^2} $. The $l_1$ norm (also called the Manhattan norm), defined as $\|x\| = \sum_{i=1}^n |x_i|$, is usually employed when we want the algorithm to select only a subset of features to learn on.

```python
# Define logistic regression hyperparameter ranges to be tested using grid search
param_grid_lr = [{'penalty':['l1','l2'],'C': np.arange(0.05,2.0,0.05)}]

# Conduct grid search for logistic regression hyperparameters
grid_lr = GridSearchCV(LogisticRegression(), param_grid_lr, 'f1', cv = 5)
grid_lr.fit(X_train_lr_, y_train)

# Print results and create a logistic regression model with chosen hyperparameters
print('Best parameters:',grid_lr.best_params_)
print('Best score:', grid_lr.best_score_)
lr = grid_lr.best_estimator_
```

```python
# Create array indicating when each penalty parameter was used during grid search
param_penalty = grid_lr.cv_results_['param_penalty']

# Create two arrays storing cross-validated F1-scores associated with each penalty type
mean_test_score_l1 = np.array(grid_lr.cv_results_['mean_test_score'][param_penalty == 'l1'])
mean_test_score_l2 = np.array(grid_lr.cv_results_['mean_test_score'][param_penalty == 'l2'])

# Create two arrays storing C parameters associated with each penalty type
param_C_l1 = np.array(grid_lr.cv_results_['param_C'][param_penalty == 'l1'])
param_C_l2 = np.array(grid_lr.cv_results_['param_C'][param_penalty == 'l2'])

# Plot cross-validated F1-score as function of C parameter for both l1 and l2 penalties
plt.figure(figsize = (5,4))
plt.plot(param_C_l1, mean_test_score_l1)
plt.plot(param_C_l2, mean_test_score_l2)
plt.legend(['l1 penalty','l2 penalty'], loc = 4)
plt.xlabel('C (inverse of regularization strength)')
plt.ylabel('F1 score')
plt.title('Optimizing parameters for logistic regression')
plt.annotate('Best parameters', xy = (1.1, .676), xytext = (1, .66),
             arrowprops=dict(facecolor='black'), color='black')
plt.show()
```

Grid search reveals that the best hyperparameters for logistic regression on its associated optimized feature set is not far from the defaults where penalty = 'l2' and C = 1.0. The tuned logistic regression model yields only a marginally better F1-score of 0.6763 over that of an untuned model, 0.6758.

### Adaboost
For [adaptive boosting](https://en.wikipedia.org/wiki/AdaBoost), we choose values for *learning_rate* and *n_estimators*. A brief overview:
- **n_estimators**: Specifies the maximum number of estimators to be used before boosting is terminated. Reducing this value is akin to increasing the regularization strength of the model.
- **learning_rate**: Specifies how much to shrink the contribution of each new classifier in the boosting series. Reducing this value is akin to increasing the regularization strength of the model. 

```python
# Define adaptive boosting hyperparameter ranges to be tested using grid search
x = np.arange(30,110,10)
y = np.arange(0.5,2.25,0.25)
param_grid_ada = [{'n_estimators': x,'learning_rate': y}]

# Conduct grid search for adaptive boosting hyperparameters
grid_ada = GridSearchCV(AdaBoostClassifier(), param_grid_ada, 'f1', cv = 5)
grid_ada.fit(X_train_ada_, y_train)

# Print results and create an adaptive boosting model with chosen hyperparameters
print('Best parameters:',grid_ada.best_params_)
print('Best score:', grid_ada.best_score_)
ada = grid_ada.best_estimator_
```

```python
# Plot cross-validated F1-score as function of learning rate and number of trees
z = grid_ada.cv_results_['mean_test_score'].reshape(len(y),len(x))

plt.figure(figsize = (5,3))
g = plt.contourf(x, y, z, cmap='plasma')
plt.colorbar(g)
plt.grid(True)
plt.ylim(1.7,2.0)
plt.title('Optimizing parameters for adaptive boosting')
plt.xlabel('Number of trees')
plt.ylabel('Learning rate')
plt.annotate('Best parameters', xy = (80, 1.75), xytext = (60, 1.82), arrowprops = {'facecolor':'white'},
             color = 'white')
plt.show()
```

From the plot above, we observe that the maximum number of estimators allowed in the boosting series does not have much effect on the effectiveness of the model, but the learning rate does.

# Evaluating individual classifiers
After using training data with optimized feature sets to tune each model, we now test how well each model performs on the test set, which is comprised entirely of previously unseen data. Because we are interested in models that provide the best F1-score rather than the greatest accuracy, we must deal with an additional parameter, the *probability threshold*.

Rather than having each classifier make a prediction about whether a respondent voted, the classifier instead outputs the *probability* that a respondent is a non-voter. This enables us to set a probability threshold, above which a respondent is classified as a non-voter, and at or below which the respondent is classified as a voter. Because different thresholds yield different F1-scores, and the threshold associated with the best F1-score is often not the same as that which yields the best accuracy, it is necessary to calculate the F1-scores for a range of thresholds in order to decide which threshold to use for the test data.

```python
def evaluate_test(y_test, y_pred, threshold = 0.5, name = 'estimator', show_curve = True):
    
    #Calculate performance metrics
    accuracy = accuracy_score(y_test, y_pred[:,1] > threshold)
    precision = precision_score(y_test, y_pred[:,1] > threshold)
    recall = recall_score(y_test, y_pred[:,1] > threshold)
    f1 = f1_score(y_test, y_pred[:,1] > threshold)
    roc = roc_auc_score(y_test, y_pred[:,1])
    
    #Print performance metrics
    print('Accuracy:', accuracy, ', Precision:', precision, ', Recall:', recall)
    print('F1 score:', f1, ', ROC area:', roc)
    print('Confusion matrix:\n',confusion_matrix(y_test, y_pred[:,1] > threshold))
    
    #Plot ROC curve
    if show_curve:
        fpr, tpr, thresholds = roc_curve(y_test, y_pred[:,1])
        plt.figure(figsize = (5,4))
        plt.plot(fpr, tpr)
        plt.xlabel('False positive rate')
        plt.ylabel('True positive rate')
        plt.title('ROC curve for {}'.format(name))
        plt.show()
        
    return [accuracy, precision, recall, f1, roc]
```

```python
def find_best_threshold(X_train, y_train, preprocessing, estimator, name = 'estimator',
                        threshold_list = np.arange(0, 1.01, .01), show_curve = True):
    
    # Train estimator and generate array of prediction probabilities
    X_train = preprocessing.fit_transform(X_train)
    estimator.fit(X_train,y_train)
    y_pred = estimator.predict_proba(X_train)
    
    # Generate array of F1-scores as function of thresholds and prints threshold that yields highest F1-score
    f1_list = []
    best_threshold = 0
    best_f1 = 0
    for threshold in threshold_list:
        f1 = f1_score(y_train, y_pred[:,1] > threshold)
        f1_list.append(f1)
        if f1 > best_f1:
            best_threshold = threshold
            best_f1 = f1
    print('Best threshold:', best_threshold, ', Best F1:', best_f1)
    
    # Plot F1-score as function of threshold
    if show_curve:
        plt.figure(figsize = (5,4))
        plt.plot(threshold_list, f1_list)
        plt.xlabel('Threshold')
        plt.ylabel('F1 score')
        plt.title('Optimizing threshold for {}'.format(name))
        plt.show()     
    return best_threshold
```

### Logistic regression

```python
threshold_lr = find_best_threshold(X_train_lr, y_train, preprocessing, lr, 'logistic regression')
```

```python
lr.fit(preprocessing.fit_transform(X_train_lr),y_train)
y_pred_lr = lr.predict_proba(preprocessing.transform(X_test_lr))
scores_lr = evaluate_test(y_test, y_pred_lr, threshold_lr, 'logistic regression', True)
```

Optimizing the threshold for logistic regression results in a respectable F1-score of **0.728** on the training data. Unfortunately, the F1-score of the classifier on test data is a significantly lower **0.595**. While not terrible, it is also not as good as hoped. Based on the [confusion matrix](http://www.dataschool.io/simple-guide-to-confusion-matrix-terminology/), at threshold of **0.34**, the model is misclassifying a roughly equal number of voters as non-voters and non-voters as voters.

```python
ada.fit(preprocessing.fit_transform(X_train_ada),y_train)
y_pred_ada = ada.predict_proba(preprocessing.transform(X_test_ada))
scores_ada = evaluate_test(y_test, y_pred_ada, threshold_ada, 'adaptive boosting', True)
```

### A note on generalizability
Here we present an example of how data from different election years may differ enough to pose significant challenges to developing generalizable models. During [feature selection](https://github.com/zhao1701/Polling-Error-Correction/blob/master/04%20Feature%20Selection.ipynb), we noticed that for logistic regression, *VCF0713_oh4*, which corresponds with whether the respondent intends to not vote, was the single most important feature from the pre-2012 data. In fact, while applying logistic regression on an optimized training subset of 115 features yielded a cross-validated F1-score of **0.67**, doing the same with *VCF0713_oh4* as the only feature produces an F1-score **0.592**. We also examined how a logistic regression model performs when the most important feature is removed; training on the remaining 114 features results in a cross-validated F1-score of **0.545**. This suggests that while *VCF0713_oh4* is a feature of significant importance and its absence from the feature set results in a drastic reduction in the effectiveness of a logistic regression model, it is not the only feature with predictive value.

We mention this because it turns out that the importance of *VCF0713_oh4* does not hold for the test set consisting of data from 2012. This can easily be observed by examining the proportion of respondents from the training set and test set who indicated an intention to not participate in the election.

```python
intend_nv_train = sum(df_orig.loc[df_orig.year < 2012,'VCF0713_oh4']) / len(df_orig[df_orig.year < 2012])
intend_nv_test = sum(df_orig.loc[df_orig.year == 2012,'VCF0713_oh4']) / len(df_orig[df_orig.year == 2012])
actual_nv_train = len(df_orig.loc[(df_orig.year < 2012) & (df_orig.final_vote == 0)]) / \
    len(df_orig[df_orig.year < 2012])
actual_nv_test = len(df_orig.loc[(df_orig.year == 2012) & (df_orig.final_vote == 0)]) / \
    len(df_orig[df_orig.year == 2012])
nv_table = np.array([[intend_nv_train, actual_nv_train],
                    [intend_nv_test, actual_nv_test]])
nv_table = pd.DataFrame(nv_table, columns = ['Proportion intended non-vote','Proportion actual non-vote'],
                        index = ['Training data (pre-2012)','Test data (2012)'])
nv_table
```

A far smaller proportion of respondents in 2012 indicated their intent to not vote in the upcoming election than in previous election years, yet the proportions of actual non-voters remains relatively the same. This means that the feature *VCF0713_oh4* is substantially less important in the test set than in the training set, which would in turn pose a significant obstacle to models seeking to generalize from one set of election years to another. Indeed, given the reduced importance of *VCF0713_oh4* in the test set, it makes sense that the performance of a model trained on pre-2012 data predicting on 2012 data is quite similar to the performance of that same model on pre-2012 data with *VCF0713_oh4* excluded (both result in F1-scores just below **0.55**).

# Soft voting classifier (part 1)
### Overview and motivation
Having evaluated each model individually, we next seek to determine whether a voting classifier might offer improved performance. Not to be confused with the problem at hand (whether or not a respondent voted), a voting classifier is an ensemble method that combines the predictions of individual models by having them "vote" on how to classify a given data point. Voting can take two forms:
1. **Hard voting**: This is the simplest voting method. Each model "votes" by making a prediction about how to classify a data point, and the classification category with the most votes wins, becoming the output of the voting classifier itself.
2. **Soft voting**: Whereas in hard voting the individual models output predictions, in soft voting each model outputs probabilities that a data point belongs to a certain classification category. The weighted average of these probabilities then determines how the voting classifier classifies a data point.
In both cases, the motivation behind using a voting classifier lies in the fact that different models have different strengths and may perform better on different subsets of data points. Combining the various strengths of different models, then, would theoretically offer improved performance.

### Custom classifier
While Scikit-learn offers its own voting classifier, it comes with two limitations:
1. Only one feature subset can be used. Therefore, out of the four feature subsets that were created, each one an optimized subset for a particular model, we would have to choose one for all individual models of a voting classifier to train on.
2. The weights for soft voting are user-specified, so the voting classifier itself does not find optimized weights itself.

To overcome these limitations, we build a custom soft voting classifier where each individual model is trained and tested on its own feature subset. To calculate a weighted average of prediction probabilities, instead of using user-specified weights, the custom soft voting classifier learns optimal weights by performing logistic regression on the probabilities output by individual models. In order to mitigate the effects of overfitting with this final layer of logistic regression, a [bagging classifier](https://en.wikipedia.org/wiki/Bootstrap_aggregating), using logistic regression as a base estimator, is used. In bagging, short for bootstrap aggregating, a number of new training sets are generated by resampling with replacement from the original training set. This allows logistic regression to be trained on different data sets, the associated coefficients of which are then averaged together, resulting in a model less at risk for overfitting than a single logistic regression.

```python
class SoftVotingClassifier:
    def __init__(self, estimators, print_cv = True):
        self.estimators = estimators
        self.n = len(estimators)
        self.blending = BaggingClassifier(LogisticRegression(), n_estimators = 20)
        self.print_cv = print_cv
        
    def fit(self, X_train_list, y_train):
        assert(self.n == len(X_train_list))
        predict_probas = np.zeros([len(y_train), self.n])
        for i in range(0, self.n):
            self.estimators[i].fit(X_train_list[i], y_train)
            predict_probas[:,i] = self.estimators[i].predict_proba(X_train_list[i])[:,1]
            
        self.blending.fit(predict_probas, y_train)
        y_pred = self.blending.predict_proba(predict_probas)[:,1]
        
        accuracy_cv = cross_val_score(self.blending, predict_probas, y_train, cv = 5, scoring = 'accuracy')
        print('Cross-validated accuracy:', accuracy_cv.mean())
        f1_cv = cross_val_score(self.blending, predict_probas, y_train, cv = 5, scoring = 'f1')
        print('Cross-validated f1:', f1_cv.mean())
        
        return y_pred
    
    def predict_proba(self, X_test_list):
        assert(self.n == len(X_train_list))
        predict_probas = np.zeros([len(y_test), self.n])
        for i in range(0, self.n):
            predict_probas[:,i] = self.estimators[i].predict_proba(X_test_list[i])[:,1]
        
        y_pred = self.blending.predict_proba(predict_probas)[:,1]
        
        return y_pred
```

```python
classifiers = [lr,ada,bnb,svm]
voting = SoftVotingClassifier(classifiers)
X_train_list = [X_train_lr, X_train_ada, X_train_bnb, X_train_svm]
X_test_list = [X_test_lr, X_test_ada, X_test_bnb, X_test_svm]
norm = Normalizer()
for i in range(0, len(X_train_list)):
    X_train_list[i] = imp.fit_transform(X_train_list[i])
    X_train_list[i] = norm.fit_transform(X_train_list[i])
    X_test_list[i] = imp.transform(X_test_list[i])
    X_test_list[i] = norm.transform(X_test_list[i])
y_pred = voting.fit(X_train_list, y_train)
```

We see that the accuracy and F1-score of the custom soft voting classifier is pretty good on the training data, though it performs only marginally better than using a support vector machine at a threshold of **0.5**.

```python
f1_list = []
best_threshold = 0
best_f1 = 0
for threshold in np.arange(0,1.01,0.01):
    f1 = f1_score(y_train, y_pred > threshold)
    f1_list.append(f1)
    if f1 > best_f1:
        best_threshold = threshold
        best_f1 = f1
print('Best threshold:', best_threshold, ', Best F1:', best_f1)
plt.figure(figsize = (5,4))
plt.plot(np.arange(0,1.01,0.01), f1_list)
plt.xlabel('Threshold')
plt.ylabel('F1 score')
plt.title('Optimizing threshold for custom soft voting classifier')
plt.show()  
```

```python
y_test_pred = voting.predict_proba(X_test_list)
```

```python
print('Test accuracy:', accuracy_score(y_test, y_test_pred > 0.1))
print('Test f1-score:', f1_score(y_test, y_test_pred > 0.1))
print('Test confusion matrix:\n', confusion_matrix(y_test, y_test_pred > 0.1))
```

Furthermore, we see that optimizing the threshold using the predicted probabilites of training data does not work particularly well, resulting in very poor performance both in terms of accuracy and F1.

# Soft voting classifier (part 2)
Although the voting classifier provided by Scikit-learn comes with certain limitations, as detailed previously, these limitations might also be what prevents the overfitting seen in the previous classifier. Using Scikit-learn's classifier, we must choose one common feature set for all individual models to train and test on. Because the logistic regression model has shown the most promise, we choose the feature set that was optimized for logistic regression.

The Scikit-learn voting classifier also requires the user to specify the weights to be applied to the results of each individual model. To find the ideal weights, we use a grid-search-like process that tests different combinations of weights.

**Note**: In evaluating the effectiveness of each weight combination, using a cross-validated F1-score should be preferred. But given the time needed to run such a procedure, we elect to use a simpler grid-search without cross-validation.

```python
best_score = 0
best_weights = None

weight_combos = itertools.product([1,2,3,4],[1,2],[1,2],[1,2,3])
for weights in weight_combos:
    voting = VotingClassifier([('lr',lr,),('ada',ada), ('bnb',bnb), ('svc',svm)], voting = 'soft',
                              weights = weights)
    X_train_vote = preprocessing.fit_transform(X_train_lr)
    voting.fit(X_train_vote, y_train)
    y_vote_pred = voting.predict(X_train_vote)
    score = f1_score(y_train, y_vote_pred)
    print('{} -> f1 score: {}'.format(weights, score))
    if score > best_score:
        best_score = score
        best_weights = weights

print('Best f1:', best_score)
print('Best weights:', best_weights)
```

```python
voting = VotingClassifier([('lr',lr,),('ada',ada), ('bnb',bnb), ('svc',svm)], voting = 'soft',
                              weights = best_weights)

X_train_vote = preprocessing.fit_transform(X_train_lr)
voting.fit(X_train_vote, y_train)
y_vote_pred = voting.predict_proba(X_train_vote)
```

The ideal weights of **(1, 1, 1, 3)** produce an F1-score of **0.766**. As with before, we optimize the threshold for further improvements to the F1-score.

```python
f1_list = []
best_threshold = 0
best_f1 = 0
for threshold in np.arange(0,1.1,0.1):
    f1 = f1_score(y_train, y_vote_pred[:,1] > threshold)
    f1_list.append(f1)
    if f1 > best_f1:
        best_threshold = threshold
        best_f1 = f1
print('Best threshold:', best_threshold, ', Best F1:', best_f1)
plt.figure(figsize = (5,4))
plt.plot(np.arange(0,1.1,0.1), f1_list)
plt.xlabel('Threshold')
plt.ylabel('F1 score')
plt.title('Optimizing threshold for soft voting classifier')
plt.show()  
```

```python
X_test_vote = preprocessing.fit_transform(X_test_lr)
y_vote_pred_test = voting.predict_proba(X_test_vote)
print('Test f1-score:', f1_score(y_test, y_vote_pred_test[:,1] > 0.4))
print('Test confusion matrix:\n', confusion_matrix(y_test, y_vote_pred_test[:,1] > 0.4))
```

Applying the voting classifier to 2012 test data using the optimized threshold of **0.4** results in an F1-score of **0.611**, which is an improvement over **0.595**, the test F1-score for logistic regression.
